```python
# import libraries
import qrcode
import matplotlib.pyplot as plt
```

```python
# add data which you want to be QR representation
data="https://youtu.be/QLKw1Jcf-FA?si=6-ewklLM7AibIbV1"
# output file name
filename="my_qr_code_.png"
#generate qr code
image=qrcode.make(data)
# save image in to a file
image.save(filename)
plt.imshow(image,cmap='gray')
```
